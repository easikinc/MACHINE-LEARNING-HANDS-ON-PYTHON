```python
cat_var=data.select_dtypes(include=['O'])
```

```python
cat_var.head(10)
```

```python
error_pct=cat_var.isnull().sum()/cat_var.shape[0]*100
```

```python
error_key=error_pct[error_pct>50].keys()
```

```python
d_cleaned_data=cat_var.drop(error_key,axis=1)
```

```python
d_cleaned_data
```

```python
missing_var=d_cleaned_data.isnull().mean()*100
```

```python
missing_var
```

```python
# select which have 0% or more error or impurities in your cols
miss_keys=missing_var[missing_var>0].keys()
```

```python
miss_keys
d_cleaned_data['time'].isnull().sum()
```

```python
time_cleaned_data=d_cleaned_data['time'].fillna(d_cleaned_data['time'].mode()[0])
```

```python
d_cleaned_data['time'].isnull().sum()
```

```python
# only time column have cleaned now we cleaned all columns by the loop
for var in miss_keys:
    d_cleaned_data[var].fillna(d_cleaned_data[var].mode()[0],inplace=True)
```
