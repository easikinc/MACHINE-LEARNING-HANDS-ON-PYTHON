# Handling missing values manually , Global constants , measure of central tendency : BY SKLEARN

```python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.impute import SimpleImputer
```

```python
# load 2 datasets 1 is Train_df   and 2nd is test_df but all datasets soud be same
train=pd.read_csv('/home/utkarss/MACHINE LEARNING HANDS ON PYTHON/DataSheets/NFL_Train.csv',low_memory=False)
test=pd.read_csv('/home/utkarss/MACHINE LEARNING HANDS ON PYTHON/DataSheets/NFL_Test.csv',low_memory=False)
```

# Now Start Numerical Imputaion ->

```python
num_vars=train.select_dtypes(include=['int64','float64']).columns
```

```python
train.isnull().sum()
```

```python
imputer_mean=SimpleImputer(strategy='mean')
imputer_mean.fit(train[num_vars])
```

```python
imputer_mean.statistics_
```

```python
# Change this array into DF
train[num_vars]=imputer_mean.transform(train[num_vars])
```

```python
train[num_vars].isnull().sum()
```

```python
# now modify the test data 
test[num_vars]=imputer_mean.transform(test[num_vars])
```

```python
train[num_vars].isnull().sum()
```

```python
test[num_vars].isnull().sum()
```

```python
# all done You can change your strategy in simple imputer method : change value by mean , median , most frequent
```

# Now Start Categorical Imputation 

```python
cat_vars=train.select_dtypes(include=['O']).columns
```

```python
cat_vars
```

```python
imputer_mode=SimpleImputer(strategy='most_frequent')
imputer_mode.fit(train[cat_vars])
```

```python
train[cat_vars].isnull().sum()
```

```python
train[cat_vars]=imputer_mode.transform(train[cat_vars])
train[cat_vars].isnull().sum()
```

```python
# change into Test dataset
test[cat_vars]=imputer_mode.transform(test[cat_vars])
```

```python
test[cat_vars].isnull().sum()
```
