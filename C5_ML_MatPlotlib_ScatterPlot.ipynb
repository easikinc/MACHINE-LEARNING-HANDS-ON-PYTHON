```python
from matplotlib import style
import matplotlib.pyplot as plt
import numpy as np 
import random
import pandas as pd
```

```python
# Scatter Plot Of Our Own Data 
data = pd.read_csv('/home/utkarss/MACHINE LEARNING HANDS ON PYTHON/DataSheets/songs.csv')
```

```python
data
```

```python
x=data.cluster
y=data.energy
```

```python
# Scatter Plot Of Our Own Data
plt.figure(figsize=(16,9))
plt.title('Spotify Songs Cluster Vs Energy', fontsize=15)
plt.scatter(x,y,c='r',alpha=1,marker='*')
plt.xlabel('Cluster')
plt.ylabel('Energy')
plt.show()
```

```python
# Scatter Plot Of Our Own Data
plt.figure(figsize=(16,9))
plt.title('Spotify Songs key Vs Energy', fontsize=15)
plt.scatter(data['key'],y,c='b',alpha=1,marker='*')
plt.xlabel('key')
plt.ylabel('Energy')
plt.show()
```

```python
plt.figure(figsize=(16,9))
plt.title('Spotify Songs Cluster Vs Energy', fontsize=15)
plt.scatter(x,y,c='r',alpha=1,marker='*',label='Cluster')
plt.scatter(data['key'],y,c='g',alpha=1,marker='^',label='Keys')
plt.xlabel('Cluster / Keys')
plt.ylabel('Energy')
plt.legend()
plt.show()
```
