```python
from scipy.stats import norm
tips_data=sns.load_dataset("tips")
sns.distplot(tips_data['total_bill'],fit=norm)
plt.show()
```

```python
from scipy.stats import norm
plt.figure(figsize=(16,9))
tips_data=sns.load_dataset("tips")
sns.distplot(tips_data['total_bill'],fit=norm,
            fit_kws={
                'color':'r',
                'linewidth':5,
                'linestyle':'--'
            },
            hist_kws ={
                'color':'g',
                'linewidth':3,
#                 'linestyle':'---'
            },
             rug=True,
             
             rug_kws= {
                 'color':'b',
                 'linewidth':2,
             },
             kde_kws={
                 'color':'g',
                 'linestyle':'--',
                 'linewidth':6
             }
            )
plt.show()
```
