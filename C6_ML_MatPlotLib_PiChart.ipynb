```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Classes=['IT','CS','Elx','civil','Machanical']
students=[57,77,67,76,34,]
plt.pie([1])
plt.show()
```

```python
plt.pie(students)
plt.show()
```

```python
plt.pie(students,labels=Classes,autopct='%0.1F%%')
plt.show()
```

```python
plt.pie(students,labels=Classes,autopct='%0.1F%%', explode=[0.1,0.3,0,0.1,0])
plt.show()
```

```python
plt.pie(students,labels=Classes,autopct='%0.1F%%', explode=[0.1,0.3,0,0.1,0],colors=['c','m','y','b','g'])
plt.show()
```

```python
plt.pie(students,labels=Classes,autopct='%0.1F%%', explode=[0.1,0.3,0,0.1,0],colors=['c','m','y','b','g'],
        shadow=True,
       radius=2)
plt.show()
```

```python
plt.pie(
    students,
    labels=Classes,
    autopct='%0.1F%%',
    explode=[0.1,0.3,0,0.1,0],
    colors=['c','m','y','b','g'],
    shadow=True,
    radius=2,
    textprops={'fontsize':16}
)
plt.show()
```

```python
plt.pie(
    students,
    labels=Classes,
    autopct='%0.1F%%',
#     explode=[0.1,0.3,0,0.1,0],
    colors=['c','m','y','b','g'],
    shadow=True,
    textprops={'fontsize':16},
#     pctdistance=2.3
#     frame=True
#     rotatelabels=True
    wedgeprops={'linewidth':4,'width':1,'edgecolor':'k'}
   
)
plt.show()
```

```python
plt.pie(
    students,
    labels=Classes,
    autopct='%0.1F%%',
    explode=[0.1,0.3,0,0.1,0],
    colors=['c','m','y','b','g'],
    shadow=True,
    textprops={'fontsize':16},
#     pctdistance=2.3
#     frame=True
#     rotatelabels=True
    wedgeprops={'linewidth':4,'width':2,'edgecolor':'k'}
   
)
plt.show()
```

```python
plt.pie(
    students,
    labels=Classes,
    autopct='%0.1F%%',
    explode=[0.1,0.3,0,0.1,0],
    colors=['c','m','y','b','g'],
    shadow=True,
    textprops={'fontsize':16},
#     pctdistance=2.3
#     frame=True
#     rotatelabels=True
    wedgeprops={'linewidth':4,'width':3,'edgecolor':'k'}
   
)
plt.show()
```

```python
plt.pie(
    students,
    labels=Classes,
    autopct='%0.1F%%',
    explode=[0.1,0.3,0,0.1,0],
    colors=['c','m','y','b','g'],
    shadow=True,
    textprops={'fontsize':16},
#     pctdistance=2.3
#     frame=True
#     rotatelabels=True
    wedgeprops={'linewidth':4,'width':1,'edgecolor':'b'}
   
)
plt.legend()
plt.show()
```
