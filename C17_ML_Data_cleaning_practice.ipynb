```python
dummy_df=pd.get_dummies(data,drop_first=True,dtype='int64')
```

```python
# Now we can see the all categorical variables ha been transformatted into numerics and column : area has large value i think i use feature scalling for it
# we use standerd scaller here to scale the all values 
from sklearn.preprocessing import StandardScaler
```

```python
#now split the column from the dataset for scale
X_metrics=dummy_df.drop(columns=['price'])
X_columns=X_metrics.columns
y=dummy_df['price']
```

```python
from sklearn.model_selection import train_test_split
X_train , X_test , Ytrain ,Ytest=train_test_split(X_metrics,y,test_size=0.2,random_state=3)
```

```python
sc=StandardScaler()
sc.fit(X_train)
```

```python
X_transformed_train=sc.transform(X_train)
X_transformed_test=sc.transform(X_test)
```

```python
Data_for_model_train=pd.DataFrame(X_transformed_train,columns=X_columns)
Data_for_model_test=pd.DataFrame(X_transformed_test,columns=X_columns)
```
