```python
#Pandas merging Function for merging dataframes
df1=pd.DataFrame({
    'Id':[1,2,3,4,5,6,7],
    'Class':[1,4,2,3,2,1,3]
})
df2=pd.DataFrame({
    'Id':[1,2,3,4,5,6,7],
    'Name':['Jayant','Nagesh','Utkarsh','Bhoopendra','Umesh','ram','Shilu']
})
merged_data=pd.merge(df1,df2)
merged_data
```
