# Matplotlib to Data Showing with numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_axis=np.arange(0,3*np.pi,0.5)
x_axis
```

```python
y_axis=np.sin(x_axis)
y_axis
```

```python
# Now printing the Curve of the Sin Theeta
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# COS Curve 
y_axis=np.cos(x_axis)
y_axis
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```

```python
# TAN Theeta's Curve
y_axis=np.tan(x_axis)
y_axis
```

```python
plt.plot(x_axis,y_axis)
plt.show()
```
