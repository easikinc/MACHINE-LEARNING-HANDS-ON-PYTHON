```python
x=np.arange(2024,2030)
y=[0,20,56,28,90,76]
plt.plot(x,y)
plt.title('Delhi temprature',fontsize=13)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Temprature',fontsize=15)
plt.axis(xmin=2023,ymin=0)
plt.show()
```

```python
#showing legend in plot 
x=np.arange(2024,2030)
y=[0,20,56,28,90,76]
plt.plot(x,y)
plt.title('Delhi temprature',fontsize=13)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Temprature',fontsize=15)
plt.axis(xmin=2023,ymin=0)
plt.legend('Temprature Line',loc=5)
plt.show()
```

```python
#Showing two lines in a one plot
x=np.arange(2024,2030)
y=[0,20,56,28,90,76]
z=[45,76,23,55,22,59]
plt.plot(x,y,color='b',marker='o',linestyle='--',linewidth=4,markersize=16 , label='Delhi')
plt.plot(x,z,color='g',marker='o',linestyle='--',linewidth=4,markersize=16 , label='Mumbai')
plt.title('Delhi temprature',fontsize=13)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Temprature',fontsize=15)
plt.axis(xmin=2023,ymin=0)
plt.legend(loc=2)
plt.show()
```

```python
#Showing two lines in a one plot
x=np.arange(2024,2030)
y=[0,20,56,28,90,76]
z=[45,76,23,55,22,59]
plt.plot(x,y,color='b',marker='o',linestyle='--',linewidth=1,markersize=5 , label='Delhi')
plt.plot(x,z,color='g',marker='o',linestyle='--',linewidth=1,markersize=5 , label='Mumbai')
plt.title('Delhi temprature',fontsize=13)
plt.xlabel('Year',fontsize=15)
plt.ylabel('Temprature',fontsize=15)
plt.axis(xmin=2023,ymin=0)
plt.legend(loc=2)
plt.show()
```

# 2. Histogram Plot

```python
# now i want more random data
import random
Boys_age=np.random.randint(10,70,50)
Girls_age=np.random.randint(18,50,50)
```

```python
plt.hist([Boys_age],label=['Boys'])
plt.title('Avg Age Of Boys')
plt.xlabel('Age')
plt.ylabel('Number Of Boys')
plt.legend(loc=1)
plt.show()
```

```python
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'])
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],histtype='step')
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],histtype='barstacked')
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# histype
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],histtype='stepfilled')
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# bins
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],bins=11)
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# rwidth

plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],rwidth=0.6)
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# align
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],align='left')
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],align='right')
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],align='mid')
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# orientation 
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],orientation='horizontal')
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],color=['green','yellow'])
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# changing background
from matplotlib import style as css
css.use('ggplot')
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],color=['green','yellow'])
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# Changing size of plot by figure()
from matplotlib import style as css
css.use('ggplot')
plt.figure(figsize=(16,9))
plt.hist([Boys_age,Girls_age],label=['Boys','Girls'],color=['green','yellow'])
plt.title('Avg Age Of Boys And Girls')
plt.xlabel('Age')
plt.ylabel('Number Of Boys And Girls')
plt.legend(loc=1)
plt.show()
```

```python
# Printing Data of A collage of three branches  CS IT ELX
CS=[12,33,55,77,88]
Year=['2019','2020','2021','2022','2023']
IT=[56,54,45,55,46]
Elx=[78,89,34,34,24]
plt.figure(figsize=(16,9))
plt.title('Collage of Technology')

plt.bar(Year,CS, color='red', width=0.8,label='CS')
plt.bar(Year,IT ,color='green', width=0.8,label='IT')
plt.bar(Year,Elx,color='yellow', width=0.8,label='Elx')
plt.xlabel(' Year')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```

```python
# Printing Data of A collage of three branches  CS IT ELX WITH SEPERATE STICKS

CS=[12,33,55,77,88]
Year=['2019','2020','2021','2022','2023']
IT=[56,54,45,55,46]
Elx=[78,89,34,34,24]
year_index=np.arange(len(Year))
width=0.2
plt.figure(figsize=(16,9))
plt.title('Collage of Technology')

plt.bar(year_index,CS,width,label='CSE',color='red',)
plt.bar(year_index+width,IT , width,label='IT',color='green',)
plt.bar(year_index+(width*2),Elx, width,label='Elx',color='yellow')
plt.xticks(year_index+width,Year, rotation=20)
plt.xlabel(' Year')
plt.ylabel('Number of Students')
plt.legend()
plt.show()
```
